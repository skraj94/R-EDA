```python
import pandas as pd
import numpy as np
ff = pd.read_csv('forestfires.csv')
ff.head(5)
```

```python
area = ff.area
deciles = np.arange(0.1,1.0,.1)
percen = []
for i in deciles:
    percen.append(np.mean(ff.area) * i)
percen
```

```python
ff['Risk1'] = ''
l =[]
for i in ff.area:
    if i > percen[8]:
        l.append ('Group_8')
    elif i < percen[7] and i > percen[6]:
        ff.Risk1[i] = 'Group_7'
    elif i < percen[6] and i > percen[5]:
        ff.Risk1[i] = 'Group_6'
    elif i < percen[5] and i > percen[4]:
        ff.Risk1[i] = 'Group_5'
    elif i < percen[4] and i > percen[3]:
        ff.Risk1[i] = 'Group_4'
    elif i < percen[3] and i > percen[2]:
        ff.Risk1[i] = 'Group_3'
    elif i < percen[2] and i > percen[1]:
        ff.Risk1[i] = 'Group_2'
    elif i < percen[1] and i > percen[0]:
        ff.Risk1[i] = 'Group_1'
    else:
        ff.Risk1[i] = 'Group_0'
ff.head(5)
l
```

```python
l =[]
for i in ff.area:
    if i > percen[8]:
        l.append ('Group_8')
    elif i < percen[7] and i > percen[6]:
        l.append ('Group_7')
    elif i < percen[6] and i > percen[5]:
        l.append ('Group_6')
    elif i < percen[5] and i > percen[4]:
        l.append ('Group_5')
    elif i < percen[4] and i > percen[3]:
        l.append ('Group_4')
    elif i < percen[3] and i > percen[2]:
        l.append ('Group_3')
    elif i < percen[2] and i > percen[1]:
        l.append ('Group_2')
    elif i < percen[1] and i > percen[0]:
        l.append ('Group_1')
    else:
        l.append ('Group_0')
print(len(l))
```

```python
grouping = pd.DataFrame(ff[['area','Risk1']],columns=['area','Risk1'])
```

```python
r1 = grouping.groupby(['Risk1']).median()['area']
#choosed median because the variable is right skewed i.e mode<median<mean
r1.index
```

```python
import matplotlib.pyplot as plt
plt.plot(r1)
plt.show()
```

```python
## From the plot we can split the groups into three low_risk, medium_risk, High_risk 
```

```python
r2 =[]
for i in grouping.Risk1:
    for j in r1.index[0:4]:
        if i == j:
            r2.append("LOW_RISK")
    for m in r1.index[4:8]:
        if i == m:
            r2.append('MEDIUM_RISK')
    for k in r1.index[8]:
        if i == k:
            r2.append('HIGH_RISK')
len(r2)
#len(grouping.Risk1)
```

```python
for i in r1.index[0:4]:
    grouping['risk'][grouping.Risk1 == i] = 'Low_risk'

```

```python
for j in r1.index[4:8]:
    grouping['risk'][grouping.Risk1 == j] = 'Medium_risk'
```

```python
grouping['risk'][grouping.Risk1 == ''] = 'High_risk'
    
```
