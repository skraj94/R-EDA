```python
# creating a new columns of ones 
radio['value'] = np.ones(radio.shape[0],dtype=int)
radio = radio.loc[:,['user','artist','value']].copy()
```

```python
# Droping the duplicate entries 
radio = radio.drop_duplicates(subset='user', keep='first')
radio = radio.reset_index().drop_duplicates(subset='index',keep='first').set_index('index')
```

```python
# creating a Association DataFrame from Pandas DataFrame using pivot 
transformed_radio = radio.pivot(index='user', columns='artist')['value'].reset_index()
transformed_radio.columns.name = None
```

```python
# mlxtend package requires high performance machine for computing large Datasets so we go with Apyori package 
```

```python
# Importing apriori function from apyori package 
from apyori import apriori
```

```python
# reading the original dataset 
radio1 = pd.read_csv("lastfm.csv")
radio1.head()
```

```python
# Finding the unique values in the Data Frame 
radio1.nunique()
```

```python
# Apyori package requires the transaction details as list. So let us first group the artists names for every user
transformed_df1 = radio1[['user','artist']].groupby('user').aggregate(lambda x: x.unique().tolist()).reset_index()
transformed_df1.head()
```

```python
# converting the Artists Name from DataFrame to list
artist_list = transformed_df1.artist.tolist()
artist_list[1:3]
```

```python
# Creating a Association rule using support = 0.01 and confidence = 0.5
association_rules1 = apriori(artist_list, min_support=0.01,min_confidence=0.5)
association_rules1 = list(association_rules1)
association_rules1[1]
```

```python
# Checking the format of association output to retrive the values and store it as dataframe for further computation and ease of calculation 
association_rules1[1][2]
```

```python
# function to store the calculated support, confidence, lift as a DataFrame
def association_DataFrame(rules):
        
    output = []

    for i in rules:

        ordered_statistics = i[2]

        for j in ordered_statistics:
            row = []

            row.append(list(j[0]))
            row.append(list(j[1]))
            row.append(round(i[1],4))
            row.append(round(j[2],4))
            row.append(round(j[3],4))

            output.append(row)

    return pd.DataFrame(output,columns=['LHS','RHS','Support','Confidence','Lift'])
```

```python
association_DataFrame(association_rules1)
```

```python
# Rule-2 with min_support = 0.015 and min_confidence = 0.5 and for three artist combinaation 
association_rules2 = apriori(artist_list, min_support=0.015,min_confidence=0.5,min_length = 3)
association_rules2 = list(association_rules2)
```

```python
association_rules2[1]
```

```python
# sorting the values of rules based on descending order of Lift value 
association_DataFrame(association_rules2).sort_values('Lift',ascending=False).reset_index(drop=True)
```
